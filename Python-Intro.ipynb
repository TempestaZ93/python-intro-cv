```python
#####################################################################
###################### M  A  T  R  I  C  E  S  ######################
#####################################################################

import numpy as np

# Different methods to create matrices

I_1 = np.identity(5)
I_2 = np.eye(3, dtype = int)

O = np.asmatrix( np.ones((2, 3)) )
Z = np.asmatrix( np.zeros((3, 2), dtype = int) )
R = np.asmatrix( np.random.rand(2, 4) )
C = np.matrix  ( [[1,2,3,4], [5,6,7,8]] )

print("I_1\n", I_1)
print("I_2\n", I_2)
print("O\n", O)
print("Z\n", Z)
print("R\n", R)
print("C\n", C)


# transpose matrizes

O_t = O.transpose()
print("O_t\n", O_t)


# matrix size information

size_complete = np.size(O)
size_rows = np.size(O, 0)
size_cols = np.size(O, 1)

print("\ncomplete size of O =", size_complete)
print("rows of O =", size_rows)
print("cols of O =", size_cols)


# Submatrices

I_1_subm = I_1[0:3, 0:4]
print("\nI_1_subm\n", I_1_subm)


# Test elements

C_g_5 = C[C>5]
print("\nC_g_5\n", C_g_5)
print("C_g_5 count", np.size(C_g_5))

C_g_5_idx = np.argwhere(C > 5)
print("C_g_5 indices\n", C_g_5_idx)


# Work with matrices

R_sorted = np.sort(R.flatten());
R_sorted = R_sorted.reshape(2, 4)
print("\nR_sorted\n", R_sorted)

O_Ot = O * O_t
print("O * Ot\n", O_Ot)
```

```python
#####################################################################
##################### S  O  L  U  T  I   O  N #######################
#####################################################################
%matplotlib inline
from matplotlib import pyplot as plt
import cv2

import scipy.io as scio
import numpy as np

# EXERCISE 1:

def LoadModifySave_Mat():
    ## Load Matrix (filename: rand.mat, matrix name: mat)
    data = scio.loadmat('./resources/rand.mat')
    mat = data.get('mat')
    
    ## Call function to calculate determinante, rank and inverse
    [det, rank, inv] = calculateStats(mat)
    
    ## Display results to user
    print("det:", det)
    print("rank:", rank)
    print("inv:\n", inv)
    
    ## Save Determinante and Rank in one File
    scio.savemat('./results/mat_det_rank.mat',{'det' : det, 'rank': rank})
    
    ## Save Matrix in other File
    scio.savemat('./results/mat_inv.mat',{'inv' : inv})
    

## Define and implement function to calculate Determinante, Rank and Inverse of any Matrix HERE:

def calculateStats(mat):
    det = np.linalg.det(mat)
    rank = np.linalg.matrix_rank(mat)
    inv = np.linalg.inv(mat)
    
    return [det, rank, inv]

# EXERCISE 2:

def LoadModifySave_Image():
    ## Load Image
    img = cv2.imread('./resources/opencv.png')
    
    ## Apply Edge Filter
    edges = cv2.Canny(img,100,200)
    
    ## Show Image/s
    fig = plt.figure()
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(img)
    ax2 = fig.add_subplot(1,2,2)
    ax2.imshow(edges, cmap = 'gray')
    plt.show()
    
    ## Save Image/s of edges
    cv2.imwrite("./results/opencv_edges.png", edges)
    
LoadModifySave_Mat()
LoadModifySave_Image()
```
